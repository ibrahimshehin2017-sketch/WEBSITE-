```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation Parameters
N = int(1e5)                 # Number of bits
sigma = 0.25                 # Turbulence strength
mu = -sigma**2 / 2           # Normalize channel
Pt = 1                       # Transmit power

SNR_dB = np.arange(0,31,2)

BER_single = []
BER_egc = []
BER_weighted = []

for snr_db in SNR_dB:

    snr = 10**(snr_db/10)
    noise_var = 1/snr

    # Generate bits
    bits = np.random.randint(0,2,N)
    s = Pt * bits

    # Generate Log-normal fading for 2 receivers
    X1 = mu + sigma*np.random.randn(N)
    X2 = mu + sigma*np.random.randn(N)

    h1 = np.exp(X1)
    h2 = np.exp(X2)

    # Generate noise
    n1 = np.sqrt(noise_var)*np.random.randn(N)
    n2 = np.sqrt(noise_var)*np.random.randn(N)

    # -------------------
    # 1) Single Receiver
    # -------------------
    r_single = h1*s + n1
    threshold_single = np.mean(h1)*Pt/2
    detected_single = (r_single > threshold_single).astype(int)
    BER_single.append(np.sum(bits != detected_single)/N)

    # -------------------
    # 2) Equal Gain Combining (EGC)
    # -------------------
    r_egc = (h1*s + n1) + (h2*s + n2)
    threshold_egc = (np.mean(h1)+np.mean(h2))*Pt/2
    detected_egc = (r_egc > threshold_egc).astype(int)
    BER_egc.append(np.sum(bits != detected_egc)/N)

    # -------------------
    # 3) Proposed Weighted Combining
    # -------------------
    w1 = h1/(h1+h2+1e-10)
    w2 = h2/(h1+h2+1e-10)

    r_weighted = w1*(h1*s + n1) + w2*(h2*s + n2)
    threshold_weighted = Pt/2
    detected_weighted = (r_weighted > threshold_weighted).astype(int)
    BER_weighted.append(np.sum(bits != detected_weighted)/N)

# Plot Results
plt.semilogy(SNR_dB, BER_single, 'o-', label='Single Receiver')
plt.semilogy(SNR_dB, BER_egc, 's-', label='EGC (2 Receivers)')
plt.semilogy(SNR_dB, BER_weighted, '^-', label='Proposed Weighted ML')

plt.grid(True)
plt.xlabel("SNR (dB)")
plt.ylabel("Bit Error Rate")
plt.title("BER vs SNR Comparison under Log-Normal Turbulence")
plt.legend()
plt.show()
```
