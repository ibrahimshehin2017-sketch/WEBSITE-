```python
!pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt

N = int(1e5)
sigma = 0.25
mu = -sigma**2 / 2

# Log-normal fading
X = mu + sigma*np.random.randn(N)
h = np.exp(X)

# OOK modulation
bits = np.random.randint(0,2,N)
Pt = 1
s = Pt * bits

SNR_dB = np.arange(0,31,2)
BER = []

for snr_db in SNR_dB:
    snr = 10**(snr_db/10)
    noise_var = 1/snr
    n = np.sqrt(noise_var)*np.random.randn(N)
    r = h*s + n

    threshold = np.mean(h)*Pt/2
    detected = (r > threshold).astype(int)

    ber = np.sum(bits != detected)/N
    BER.append(ber)

plt.semilogy(SNR_dB, BER, marker='o')
plt.grid(True)
plt.xlabel("SNR (dB)")
plt.ylabel("Bit Error Rate")
plt.title("BER vs SNR for Log-Normal FSO Channel")
plt.show()
```
