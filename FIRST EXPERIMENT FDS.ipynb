```python
num = int(input("Enter an integer: "))

# Positive, Negative or Zero
if num > 0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")

# Even or Odd
if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
import math

def statistics(numbers):
    maximum = max(numbers)
    minimum = min(numbers)
    mean = sum(numbers) / len(numbers)

    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    std_dev = math.sqrt(variance)

    return maximum, minimum, mean, std_dev

# Example usage
nums = list(map(float, input("Enter numbers separated by space: ").split()))
maxi, mini, mean, std = statistics(nums)

print(f"Maximum: {maxi}")
print(f"Minimum: {mini}")
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std:.2f}")
```

```python
lst = []

n = int(input("Enter number of elements: "))
for i in range(n):
    lst.append(int(input("Enter element: ")))

print("Original List:", lst)

# Append
element = int(input("Enter element to append: "))
lst.append(element)
print("After Append:", lst)

# Remove
rem = int(input("Enter element to remove: "))
if rem in lst:
    lst.remove(rem)
print("After Remove:", lst)

# Sort
lst.sort()
print("After Sort:", lst)

# Reverse
lst.reverse()
print("After Reverse:", lst)
```

```python
sentence = input("Enter a sentence: ").lower()
words = sentence.split()

freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print("\nWord Frequency:")
for word, count in freq.items():
    print(f"{word}: {count}")
```

```python
import os

file_name = "input.txt"

# Check if input.txt exists, if not, create a dummy one
if not os.path.exists(file_name):
    print(f"'{file_name}' not found. Creating a sample '{file_name}' with dummy data.")
    with open(file_name, "w") as f:
        f.write("10 20 30 40 50")

# Read numbers from file
with open(file_name, "r") as file:
    numbers = list(map(float, file.read().split()))

total = sum(numbers)
average = total / len(numbers)

# Write results to another file
with open("output.txt", "w") as file:
    file.write(f"Sum: {total}\n")
    file.write(f"Average: {average}\n")

print("Results written to output.txt")
```

```python
import numpy as np

# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([5, 4, 3, 2, 1])

print("1D Array 1:", arr1)
print("1D Array 2:", arr2)

# Element-wise Addition
print("Element-wise Addition:", arr1 + arr2)

# Mean and Standard Deviation
print("Mean of arr1:", np.mean(arr1))
print("Standard Deviation of arr1:", np.std(arr1))

# 2D Array
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("2D Array:\n", matrix)

# Transpose
print("Transpose of Matrix:\n", matrix.T)
```

```python
import pandas as pd

# Create a dictionary of student details
data = {
    "Roll_No": [101, 102, 103, 104, 105, 106],
    "Name": ["Aisha", "Rahul", "Meera", "Arjun", "Sneha", "Kiran"],
    "Maths": [85, 78, 92, 88, 76, 95],
    "Physics": [80, 74, 89, 90, 70, 96],
    "Chemistry": [82, 79, 94, 86, 75, 93]
}

# Create DataFrame
df = pd.DataFrame(data)

print("----- Student DataFrame -----")
print(df)

# Display first five records
print("\nFirst Five Records:")
print(df.head())

# Column-wise mean (only numeric columns)
print("\nColumn-wise Mean:")
print(df.mean(numeric_only=True))

# Rows satisfying condition (Maths > 85)
print("\nStudents with Maths marks > 85:")
print(df[df["Maths"] > 85])
```
